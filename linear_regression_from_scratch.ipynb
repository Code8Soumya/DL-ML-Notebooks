```python
def line(theta0, theta1, x):
    return theta0 + (theta1*x)
```

```python
def plotLine(theta0, theta1, x, y):
	max_x = np.max(x)
	min_x = np.min(x)

	xplot = np.linspace(min_x, max_x, 100)
	yplot = theta0 + theta1 * xplot
	plt.plot(xplot, yplot, color="#ff0000", label="Regression Line")

	plt.scatter(x,y)

	plt.axis([-5, 5, -120, 120])
	plt.show()
```

```python
theta0 = np.random.uniform(-100, 100)
theta1 = np.random.uniform(-100, 100)

plotLine(theta0, theta1, x, y)
```

```python
def avgcost(theta0, theta1, x, y):
    totalcost = 0
    for (xi, yi) in zip(x, y):
        totalcost += 0.5 * ((line(theta0, theta1, xi) - yi)**2)
    avg_cost = totalcost / len(x)

    return avg_cost
```

```python
def derivatives(theta0, theta1, x, y):
	dtheta0 = 0
	dtheta1 = 0
	for (xi, yi) in zip(x, y):
		dtheta0 += line(theta0, theta1, xi) - yi
		dtheta1 += (line(theta0, theta1, xi) - yi)*xi

	dtheta0 /= len(x)
	dtheta1 /= len(x)

	return dtheta0, dtheta1
```

```python
def updateParameters(theta0, theta1, x, y, alpha):
	dtheta0, dtheta1 = derivatives(theta0, theta1, x, y)
	theta0 = theta0 - (alpha * dtheta0)
	theta1 = theta1 - (alpha * dtheta1)

	return theta0, theta1
```

```python
def LinearRegression(x, y):
	theta0 = np.random.rand()
	theta1 = np.random.rand()

	for i in range(0, 1000):
		if i % 100 == 0:
			print(avgcost(theta0, theta1, x, y))
			plotLine(theta0, theta1, x, y)

		theta0,theta1 = updateParameters(theta0, theta1, x, y, 0.005)
```
